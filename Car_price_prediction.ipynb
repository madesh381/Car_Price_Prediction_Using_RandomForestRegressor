```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df_label.head()
```

```python
df1 = df_label.drop(columns=["Manufacturer","Year of manufacture"])
df1
```

```python
x=df1.iloc[:,df1.columns!='Mileage']
y=df1.iloc[:,df1.columns=='Mileage']
```

```python
df1.head()
```
